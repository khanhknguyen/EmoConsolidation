NOTE: The purpose of this file is to visualize performance of individual participants, but these data will not be used for paper.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
droplist = [10]
results = pd.read_csv("master_results.csv")
results = results.set_index("Participant")
results = results.drop(droplist)
pt = results.index
total_pt = pt[-1]
ntrials = 48
```

### AB Overall Accuracy


```python
#AB1 Overall Accuracy
AB1allCorr = results.AB1allCorr
plt.figure()
plt.figure(figsize = (8,6))
plt.bar(pt, AB1allCorr, alpha = 0.3, color = 'k', label = "data1")
plt.xlabel("participant", fontsize = "large")
plt.ylabel("correct trial counts", fontsize = "large")
plt.ylim(0,ntrials)
plt.xticks(np.arange(0,total_pt + 1,1))
plt.yticks(np.arange(0,ntrials,5))
plt.title("AB1 Overall Accuracy")
plt.show()
```

```python
#AB2 Overall Accuracy
AB2allCorr = results.AB2allCorr
plt.figure()
plt.figure(figsize = (8,6))
plt.bar(pt, AB2allCorr, alpha = 0.3, color = 'k', label = "data1")
plt.xlabel("participant", fontsize = "large")
plt.ylabel("correct trial counts", fontsize = "large")
plt.ylim(0,ntrials)
plt.xticks(np.arange(0,total_pt + 1,1))
plt.yticks(np.arange(0,ntrials,5))
plt.title("AB2 Overall Accuracy")
plt.show()
```

```python
#AB3 Overall Accuracy
AB3allCorr = results.AB3allCorr
plt.figure()
plt.figure(figsize = (8,6))
plt.bar(pt, AB3allCorr, alpha = 0.3, color = 'k', label = "data1")
plt.xlabel("participant", fontsize = "large")
plt.ylabel("correct trial counts", fontsize = "large")
plt.ylim(0,ntrials)
plt.xticks(np.arange(0,total_pt + 1,1))
plt.yticks(np.arange(0,ntrials,5))
plt.title("AB3 Overall Accuracy")
plt.show()
```
