This takes
- a folder with the participant number (ie. p001) that contains all raw CSV files that output from PsychoPy (total of 7)
- a CSV file (raw) from Qualtrics "Questionnaires" with all numeric values

This outputs
- a CSV file specific to the participant titled "p###_results.csv" - these individual participant CSVs can be combined in master_results.ipynb to create a full dataset

```python
import numpy as np
import pandas as pd
import os
import sys
import math
from numpy.random import default_rng
```

```python
def average(lst):
    return math.fsum(lst) / len(lst)
```

### AC Test Accuracy


```python
searchString = "ACtest"

#look for AB1 test file and open csv

for filename in dir:
  path = folder + os.sep + filename
  if os.path.isfile(path):
    f = open(path, 'r')

    if searchString in f.read():
      print("found string in file %s" %filename)
      test_csv = pd.read_csv(path)
#    else:
#      print("string not found")
    f.close()

#collect columns of interest
ACcorrResp = test_csv["key_resp.corr"]
ACtrialType = test_csv["trialtype"]
ACrespTime = test_csv["key_resp.rt"]

#for iterating
length = len(ACcorrResp) #number of trials: 48

#empty lists
corrNeu = []
corrFear = []
rtNeu = []
rtFear = []

#sort
for i in range(length): 
  if ACtrialType[i] == "neutral":
    corrNeu.append(ACcorrResp[i])
    rtNeu.append(ACrespTime[i])
  if ACtrialType[i] == "fearful":
    corrFear.append(ACcorrResp[i])
    rtFear.append(ACrespTime[i])

#clean
rtAll = [i for i in ACrespTime if str(i) != 'nan']
rtNeu = [i for i in rtNeu if str(i) != 'nan']
rtFear = [i for i in rtFear if str(i) != 'nan']

#log
ACallCorr = sum(ACcorrResp)
ACneuCorr = sum(corrNeu)
ACfearCorr = sum(corrFear)
ACallRT = average(rtAll)
ACneuRT = average(rtNeu)
ACfearRT = average(rtFear)

#percent accuracy
ACallAcc = (ACneuCorr + ACfearCorr) / (len(corrNeu) + len(corrFear))
ACneuAcc  = ACneuCorr/len(corrNeu)
ACfearAcc = ACfearCorr/len(corrFear)

#create list of faces from correct trials from AB and BC
corrABface = list(corrABface)
corrBCface = list(corrBCface)

#for combining lists into new list of unique values
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
    
#intersection of lists with unique faces
corrBface = intersection(corrABface,corrBCface)

#temporary dataframe for only AC trials that people got correct in AB or BC
tempDf = test_csv.loc[test_csv['ACcorrFace'].isin(corrBface)]

#separate temp dataframe into neutral and fearful
ACneutral = tempDf.loc[tempDf['trialtype'] == 'neutral']
ACfearful = tempDf.loc[tempDf['trialtype'] == 'fearful']

#downsample
lenNeu = len(ACneutral)
lenFear = len(ACfearful)

while lenNeu != lenFear:
  if lenNeu > lenFear:
  
    #randomly drop 1 index from neutral
    arr_indices_top_drop = default_rng().choice(ACneutral.index, size = 1, replace = False)
    ACneutral.drop(index = arr_indices_top_drop, inplace=True)

    #get new lengths
    lenNeu = len(ACneutral)
    lenFear = len(ACfearful)

  if lenFear > lenNeu:

    #randomly drop 1 index from fearful
    arr_indices_top_drop = default_rng().choice(ACfearful.index, size = 1, replace = False)
    ACfearful.drop(index = arr_indices_top_drop, inplace=True)

    #get new lengths
    lenNeu = len(ACneutral)
    lenFear = len(ACfearful)

#percent accuracy adjusted for previous performance in AB/BC
adjACallAcc = (sum(ACneutral['key_resp.corr']) + sum(ACfearful['key_resp.corr'])) / (len(ACneutral['key_resp.corr']) + len(ACfearful['key_resp.corr']))
adjACneuAcc = sum(ACneutral['key_resp.corr'])/len(ACneutral['key_resp.corr'])
adjACfearAcc = sum(ACfearful['key_resp.corr'])/len(ACfearful['key_resp.corr'])
adjACLen = lenNeu
```
